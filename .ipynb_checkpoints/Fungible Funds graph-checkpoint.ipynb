## Money supply and Money demand vs Interest rate

### Economic theory:
The money market, loanable funds or fungible funds graph contains a money supply and money demand curve and describes the relationship between quantity of loanable funds (the x-axis) and the interest rate (the y-axis). In class, we learned that money supply is a vertical line if targettedby the Federal Reserve. Without Fed intervention, however, MS is an upward sloping curve equal to savings and MD is downward sloping, equal to investment. At equilibrium, the intersection of MS and MD gives the interest rate.

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
import scipy
```

```python
def read_csv(filepath):
    df = pd.read_csv(filepath)
    df['date'] = pd.to_datetime(df['Date'])
    df.drop('Date', axis=1, inplace=True)
    df['year'], df['month'] = df['date'].dt.year, df['date'].dt.month
#     df['decade'] = np.floor(df['year']/10)*10
    print filepath
    print df.head(), '\n...\t...\t...\t...\t...\n'
    return df
    
rdf = read_csv('data/us_interest_rate.csv')
Idf = read_csv('data/us_investment.csv')
M2df = read_csv('data/us_money_supply.csv')
```

```python
def rename_col(df, oldname, newname):
    if oldname in df.columns:
        df[newname] = df[oldname]
        df.drop(oldname, axis=1, inplace=True)

rename_col(rdf, 'In percent', 'r')
rename_col(Idf, 'In US dollars per year', 'investment')
rename_col(M2df, 'In US dollars', 'M2')

merged = Idf.merge(M2df, on=['year', 'month'], suffixes=('_I', '_M2'))\
            .merge(rdf, on=['year'], suffixes=('', '_r'))\
            .drop('date_I', 1)\
            .drop('date_M2', 1)\
            .drop('date', 1)

print merged.shape[0], 'data points'
merged.head()
```

**Summary statistics:**

```python
merged.drop(['month', 'month_r'], axis=1).describe()
```

### Data analysis: Money demand
Regression of US fixed private investment against real interest rate in the same time period **reveals a downward sloping money demand-interest rate graph**. This corresponds to theoretical expectations - as interest rates increase, borrowing funds to purchase capital becomes more expensive for firms, and private capital investment falls. Thus, as interest rates increase, MD falls.

```python
sns.jointplot(data=merged, x='investment', y='r', kind='reg', color='coral', ylim=(-2, 12))
sns.plt.suptitle('Money demand, 1981 to 2014:\nFixed private investment (in trillions of $) vs real interest rate (%)',
                 y=1.08, fontsize=14, fontweight='bold')
```

### Data analysis: Money supply
However, regression of US money supply, represented by M2, against real interest rate from 1981 to 2014 **also reveals a downward sloping M2-r graph**. This is the opposite of our expectation from the loanable funds graph. The downward sloping MS graph could be explained by Federal Reserve intervention - as the Fed increases the money supply manually, the interest rate tends to fall (the vertical MS line shifts to the right in the loanable funds graph).

```python
sns.jointplot(data=merged, x='M2', y='r', kind='reg', color='green', ylim=(-2, 12))
sns.plt.suptitle('Money supply (M2, in $10 trillion)\nvs real interest rate (%)\n1981 to 2014',
                 y=1.1, fontsize=14, fontweight='bold')
```
